# Functions

```python
def getBasicStat(column, df,imgSize):
    mean = df[column].mean()
    max_ = df[column].max()
    min_ = df[column].min()
    std = df[column].std()
    df = df[df[column]==max_]
    df = df[[ 'epoch', 'precision', 'recall', 'AP', 'F1']]
    print(df.to_string(index=False))
    print('column: {} image size: %d '.format(column) %(imgSize))
    print('mean: %f'%(mean))
    print('max: %f' %(max_))
    print('min: %f' %(min_))
    print('std: %f' %(std))
def returnBasicStat(column, df,imgSize):
    mean = df[column].mean()
    max_ = df[column].max()
    min_ = df[column].min()
    std = df[column].std()
    df = df[df[column]==max_]
    epoch = df[['epoch']].epoch.item()
    return imgSize,mean,max_,min_,std,epoch
```

# Precision

```python
stats = pd.DataFrame(columns=['imgSize','mean','max','min','std','epoch'])
stats.loc[0]= returnBasicStat('precision', df_832, '832')
stats.loc[1]= returnBasicStat('precision', df_608, '608')
stats.loc[2]= returnBasicStat('precision', df_480, '480')
stats.loc[3]= returnBasicStat('precision', df_clean_480, '480_clean')
stats.loc[4]= returnBasicStat('precision', df_416, '416')
stats.loc[5]= returnBasicStat('precision', df_clean_416, '416_clean')
stats.loc[6]= returnBasicStat('precision', df_2Cls_416, '2 classes')
stats.loc[7]= returnBasicStat('precision', df_one, '416 one class')
stats.loc[8]= returnBasicStat('precision', df_two,'416 two class')
stats.loc[9] = returnBasicStat('precision', df_2CLS_edited,'416 two class edited')
stats
```

# F1

```python
stats = pd.DataFrame(columns=['imgSize','mean','max','min','std','epoch'])
stats.loc[0]= returnBasicStat('F1', df_832, '832')
stats.loc[1]= returnBasicStat('F1', df_608, '608')
stats.loc[2]= returnBasicStat('F1', df_480, '480')
stats.loc[3]= returnBasicStat('F1', df_clean_480, '480_clean')
stats.loc[4]= returnBasicStat('F1', df_416, '416')
stats.loc[5]= returnBasicStat('F1', df_clean_416, '416_clean')
stats.loc[6]= returnBasicStat('F1', df_2Cls_416, '2 classes')
stats.loc[7]= returnBasicStat('F1', df_one, '416 one class')
stats.loc[8]= returnBasicStat('F1', df_two, '416 two class')
stats.loc[9] = returnBasicStat('F1', df_2CLS_edited,'416 two class edited')
stats
```

```python
def plotCols(column,df,imgSize):
    x=df['epoch']
    y=df[column]
    plt.plot(x,y)
    plt.xlabel('epoch')
    plt.ylabel(column)
    plt.title('column: {} image size: %d '.format(column) %(imgSize))
    plt.show()
```

```python
plotCols('F1',df_832,832)
plotCols('F1',df_608,608)
plotCols('F1',df_480,480)
plotCols('F1',df_416,416)
plotCols('F1',df_clean_416,416)
```
